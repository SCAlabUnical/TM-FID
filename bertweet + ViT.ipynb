```python
class CustomDataset(Dataset):
    def __init__(self, root_dir, image_folder, transform=None):
        self.root_dir = root_dir
        self.image_folder = image_folder
        self.transform = transform
        self.data = self.load_data()

    def load_data(self):
        data = []

        for folder_name in os.listdir(self.root_dir)[:12000]:
            folder_path = os.path.join(self.root_dir, folder_name)
            json_path = os.path.join(folder_path, 'tweets.json')

            if os.path.exists(json_path):
                try:
                    with open(json_path, 'r') as f:
                        tweets_data_list = json.load(f)

                    tweet_number = folder_name.split('-')[-1]
                    image_name = f'{tweet_number}.jpg'
                    image_path = os.path.join(self.image_folder, image_name)

                    if os.path.exists(image_path) and self.is_valid_image(image_path):
                        data.append({'text': tweets_data_list, 'image_path': image_path})

                except json.JSONDecodeError as json_error:
                    print(f"Error decoding JSON in file {json_path}: {json_error}")
            else:
                print(f"JSON file not found: {json_path}")

        return data

    def is_valid_image(self, file_path):
        try:
            img = Image.open(file_path)
            img.verify()
            return True
        except Exception as e:
            print(f"Invalid image file: {file_path}. Error: {e}")
            return False

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data[idx]['text']
        image_path = self.data[idx]['image_path']

        try:
            image = Image.open(image_path).convert('RGB')

            if self.transform:
                image = self.transform(image)

            return {'text': text, 'image': image}

        except OSError as e:
            print(f"Error opening image file {image_path}: {e}")
            return None
```

# Final Model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

EMB_DIM = 768
HIDDEN = 512

input_img_embs = layers.Input(shape=(EMB_DIM,), dtype=tf.float32, name='input_img')
print(input_img_embs.shape)

input_text_embs = layers.Input(shape=(EMB_DIM,), dtype=tf.float32, name='input_text')
print(input_text_embs.shape)

# # cross-attention module

# projections in low-d space
img_emb_proj = layers.Dense(units=HIDDEN, use_bias=False)(input_img_embs)
text_emb_proj = layers.Dense(units=HIDDEN, use_bias=False)(input_text_embs)

# sigmoid masks
img_emb_mask = layers.Dense(units=HIDDEN, activation="sigmoid", use_bias=False)(input_img_embs)
text_emb_mask = layers.Dense(units=HIDDEN, activation="sigmoid", use_bias=False)(input_text_embs)

# cross-attention features
cross_text_img = img_emb_proj * text_emb_mask
cross_img_text = text_emb_proj * img_emb_mask
concat = layers.Concatenate()([cross_text_img, cross_img_text])

# classification head
x = layers.Dense(HIDDEN, activation="relu")(concat)
x = layers.Dense(HIDDEN, activation="relu")(x)

probs = layers.Dense(2, activation="softmax")(x)

model = keras.Model(inputs=[input_img_embs, input_text_embs], outputs=probs)

model.summary()
```

```python
import tensorflow_addons as tfa
from keras.callbacks import ModelCheckpoint

max_epochs = 4
batch_size = 32
opt = tfa.optimizers.RectifiedAdam(learning_rate=1e-3)
loss = keras.losses.CategoricalCrossentropy()
best_weights_file = "fake_news_detection_weights_bertweet+vit.h5"
auc = keras.metrics.AUC(curve="ROC")
m_ckpt = ModelCheckpoint(best_weights_file, monitor='val_'+auc.name, mode='max', verbose=2,
                          save_weights_only=True, save_best_only=True)
model.compile(loss=loss, optimizer=opt, metrics=[auc, keras.metrics.CategoricalAccuracy()])
```

```python
# test the model (classification report)
from sklearn.metrics import classification_report
best_weights_file = "fake_news_detection_weights_bertweet+vit.h5"
model.load_weights(best_weights_file)
opt = tfa.optimizers.RectifiedAdam(learning_rate=1e-3)
model.compile(loss=loss, optimizer=opt, metrics=[auc, keras.metrics.CategoricalAccuracy()])

y_pred_probs = model.predict([test_img, test_sentences])
y_pred = np.argmax(y_pred_probs, axis=1)
labels_test_decode = np.argmax(y_test, axis=1)

report = classification_report(labels_test_decode, y_pred, digits=3)
print(report)
```
